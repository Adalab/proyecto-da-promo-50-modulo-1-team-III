1. Bienvenida

```python
def bienvenida_juego():
    saludo= ('¡¡Bienvenidas al juego del ahorcado!!').upper()
    instrucciones= ("""Las reglas del juego son las siguientes:
    1.Un jugador elige una palabra secreta y dibuja un espacio para cada letra de la palabra."
    2.El otro jugador intenta adivinar letras para completar la palabra."
    3.Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.
    4.Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo del ahorcado.
    5.El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo del muñeco ahorcado.
      Tienes 6 intentos."
                    ¿¿COMENZAMOS??""")
    print(saludo)
    print(instrucciones)
    while True:
        start= input('Escribe jugar o salir').lower()
        if start== ('jugar'):
            return desarrollo_juego()
        elif start==('salir'):
            print("""Estaremos esperando para cuando estes preparada,
                ¡NOS VEMOS PRONTO!""")
            return
        else:
            print("Opción no válida. Por favor, escribe jugar o salir")
            
```

2. Desarrollo juego

```python
def desarrollo_juego():
  # Jugador 1
  palabra_jugador1 = input("Jugador 1, ingresa una palabra: ").lower()
  # Convertir letra en _
  longitud = len(palabra_jugador1)
  espacios = longitud * " _ "
  # Jugador 2
  print(f"Jugador 2, la palabra que tienes que adivinar tiene {longitud} letras")
  print(espacios)
  print("Tienes 6 intentos")

  # Lista para guardar los errores y avisar si la letra ya se ha usado
  lista_errores = []
  # Contador para los fallos
  intentos = 7
  dibujo = """
            ------
            |    |
            |
            |
            |
            |
          -------"""
  dibujo1 = """
            ------
            |    |
            |    O
            |
            |
            |
          -------"""
  dibujo2 = """
            ------
            |    |
            |    O
            |    |
            |
            |
          -------"""
  dibujo3 = """
            ------
            |    |
            |    O
            |   /|
            |
            |
          -------"""
  dibujo4 = """
            ------
            |    |
            |    O
            |   /|\\
            |
            |
          -------"""
  dibujo5 = """
            ------
            |    |
            |    O
            |   /|\\
            |   /
            |
          -------"""
  dibujo6 = """
            ------
            |    |
            |    O
            |   /|\\
            |   / \\
            |
          -------"""

  # Hacer una lista con cada una de las letras de la palabra a adivinar
  convertir_palabra = list(palabra_jugador1)
  posiciones = []
  # Eliminar los espacios de la lista
  conversion_espacio = list(espacios.replace(" ", ""))

  # Lista para guardar los aciertos y avisar si la letra ya se ha usado
  lista_aciertos = []
  abecedario = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "ñ", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z", "á", "é", "í", "ó", "ú", "ü", "ë"]

  # Definir conversion_frase con un valor inicial
  conversion_frase = " ".join(conversion_espacio)

  while intentos > 0:
    # Crear la condición de solo ingresar letras
    while True:
      letras_jugador2 = input("Jugador 2, elige una letra: ").lower()
      if letras_jugador2 not in abecedario:
        print('No has introducido una letra del abecedario')
        continue 
            # Crear la condición para los aciertos
      if letras_jugador2 in palabra_jugador1 and letras_jugador2 not in lista_aciertos:
        lista_aciertos.append(letras_jugador2)
        print("Has acertado!")
        # Convertir _ en la letra acertada
        for i in range(len(palabra_jugador1)):
            if palabra_jugador1[i] == letras_jugador2:
                posiciones.append(i)
                conversion_espacio[i] = letras_jugador2
                # Agregar espacios entre cada caracter
        conversion_frase = " ".join(conversion_espacio)
        print(conversion_frase)

            # Crear la condición para avisar de que ya se usó esa letra
      elif letras_jugador2 in palabra_jugador1 and letras_jugador2 in lista_aciertos:
            print(f"Ya lo has intentado con la letra: {letras_jugador2}")

            # Crear la condición para el error
      elif letras_jugador2 not in palabra_jugador1 and letras_jugador2 not in lista_errores:
        lista_errores.append(letras_jugador2)
            # Contador de fallos
        intentos -= 1
        print(f"Has fallado :(. Te quedan {intentos} intentos.")
        print(f"Tus intentos anteriores {lista_errores}")
        # Añadir dibujo a cada uno de los fallos
        if intentos == 6:
            print(dibujo1)
        elif intentos == 5:
            print(dibujo2)
        elif intentos == 4:
            print(dibujo3)
        elif intentos == 3:
            print(dibujo4)
        elif intentos == 2:
            print(dibujo5)
        elif intentos == 1:
            print(dibujo6)
            print("Has perdido :(. Ya no te quedan más intentos.")
            return # Salida si pierde

        # Crear condición para avisar que se usó esa letra
      elif letras_jugador2 not in palabra_jugador1 and letras_jugador2 in lista_errores:
        print(f"Ya lo has intentado con la letra: {letras_jugador2}")

        # Crear condición independiente para avisar de que se acertó la palabra
      if conversion_frase == " ".join(palabra_jugador1):
        print("ENHORABUENA! HAS GANADO :)")
        return # Salida si gana
      break # Salida del bucle try... except
```
